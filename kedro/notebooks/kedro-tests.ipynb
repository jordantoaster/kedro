## Load Data from Catalog

```python
from kedro.io import DataCatalog
from kedro.extras.datasets.pandas import (
    CSVDataSet,
)

io = DataCatalog(
    {
        "reviews": CSVDataSet(filepath="../data/01_raw/reviews.csv")
    }
)

reviews = io.load("reviews")  # data is now loaded as a DataFrame in 'reviews'
reviews.head(5)
```
